# real answer starting from here

```python
a[' funding_total_usd '] = np.where(a[' funding_total_usd ']== '-','0',a[' funding_total_usd '])
a[' funding_total_usd '] = a[' funding_total_usd '].apply(lambda x: str(x).replace(',',''))
a[' funding_total_usd '] = a[' funding_total_usd '].astype('float')
a.groupby(['country_code'])[' funding_total_usd '].sum().reset_index()
```

```python
Above results shows that, four steps involved to return the total funding.
1. convert special character in white space
2.convert string to float
3.return the result with the help of groupby comment
```
