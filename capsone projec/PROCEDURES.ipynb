```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(file))
file = file[(z<3).all(axis = 1)]
```

```python
f,ax = plt.subplots(figsize = (20,20))
sns.heatmap(file.corr(),annot =True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
file_scaler = scaler.fit_transform(file)
```

```python
file_scaler
```

```python
from sklearn.cluster import KMeans
scores_1 = []
lim = range(1,20)
for i in lim:
    cluster = KMeans(n_clusters = i)
    cluster.fit(file_scaler)
    scores_1.append(cluster.inertia_)
print(scores_1)    
```

```python
plt.plot(lim,scores_1)
plt.xlabel("NO of clusters")
plt.ylabel("Inertia values")
plt.xticks(np.arange(1, 20))
```

```python
kmeans = KMeans(6)
kmeans.fit(file_scaler)
labels = kmeans.labels_
```

```python
labels_act = pd.DataFrame(labels, columns = ['Clusters'])
labels_act
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = file.columns)
cluster_centers
```

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers, columns = file.columns)
cluster_centers
```
