# Standardatation step

```python
data1 = data.iloc[:,0:6]
data1.drop(['Attrition','BusinessTravel','Department'],axis = 1,inplace = True)
from sklearn import preprocessing
dataset1_std = preprocessing.scale(data1)
dataset1_std = pd.DataFrame(dataset1_std)
dataset1_std
```

# WSS PLOT or Elbow curve to findout the optimium cluster identification

```python
import matplotlib.pyplot as plt  
import pandas as pd  
%matplotlib inline
import numpy as np
plt.figure(figsize=(10,8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 1000)
    kmeans.fit(dataset1_std)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# K-Means Clustering & Cluster Information

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 1000)
y_kmeans = kmeans.fit_predict(dataset1_std)
#beginning of  the cluster numbering with 1 instead of 0
y_kmeans1=y_kmeans
y_kmeans1=y_kmeans+1
# New Dataframe called cluster
cluster = pd.DataFrame(y_kmeans1)
# Adding cluster to the Dataset1
data1['cluster'] = cluster
#Mean of clusters
kmeans_mean_cluster = pd.DataFrame(round(data1.groupby('cluster').mean(),1))
kmeans_mean_cluster
```
